## 1. Introduction to Linear Algebra <a name="intro"></a>

**Linear algebra** is the mathematical framework that supports nearly all of modern science and engineering. It is the study of vectors and the rules for manipulating them through linear transformations. In the context of Machine Learning, it is the primary language used to describe data, models, and algorithms.

### Why It Matters

- **Data Representation**: Every digital image, audio clip, or tabular data point is essentially a vector or a matrix.
- **Neural Networks**: Deep learning models are compositions of linear operations (matrix multiplications) and non-linear activations.
- **Dimensionality Reduction**: Techniques like PCA rely on the geometric properties of vectors to simplify complex datasets.

### The Principle of Linearity

A function $f$ is **linear** if it satisfies two properties:
1. **Additivity**: $f(\mathbf{v} + \mathbf{w}) = f(\mathbf{v}) + f(\mathbf{w})$
2. **Homogeneity**: $f(c\mathbf{v}) = c f(\mathbf{v})$

These properties ensure that the transformation preserves the "grid-like" structure of space.

## 2. Vectors: Foundations <a name="vectors"></a>

A **vector** is an object that has both magnitude and direction. Geometrically, we visualize it as an arrow starting from the origin $(0,0)$. Algebraically, it is an ordered list of numbers representing coordinates.

### Notation
$$\mathbf{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}$$

### Magnitude (L2 Norm)
The length of a vector in Euclidean space is calculated using the Pythagorean theorem generalization:
$$\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2}$$


```python
import numpy as np

v = np.array([3, 4])
magnitude = np.linalg.norm(v)
print(f"Vector: {v}")
print(f"Magnitude: {magnitude}")
```

## 3. Vector Operations <a name="operations"></a>

We can manipulate vectors through addition and scalar multiplication. These operations form the basis of **linear combinations**.

```python
w = np.array([1, 2])
v_plus_w = v + w
v_scaled = 2 * v
print(f"Addition: {v_plus_w}")
print(f"Scaling: {v_scaled}")
```

## 4. Dot Product and Projections <a name="projections"></a>

The **Dot Product** $(\mathbf{v} \cdot \mathbf{w})$ measures how much two vectors point in the same direction. It is a single scalar value.

### Geometric Interpretation
$$\mathbf{v} \cdot \mathbf{w} = \|\mathbf{v}\| \|\mathbf{w}\| \cos(\theta)$$

If the dot product is zero, the vectors are **orthogonal** (at a 90-degree angle).

### Projections
A **projection** is like the shadow of one vector onto another. It is a fundamental operation in finding the "closest" approximation of one data point relative to a specific axis.

```python
dot_product = np.dot(v, w)
print(f"Dot Product: {dot_product}")

# Projection of v onto w
proj_v_w = (np.dot(v, w) / np.dot(w, w)) * w
print(f"Projection of v onto w: {proj_v_w}")
```

## 5. Orthogonality <a name="orthogonality"></a>

Two vectors are **orthogonal** if their dot product is exactly zero. In high-dimensional space, orthogonality is a powerful property because it implies independence: moving along one vector has no effect on your position relative to the other.

```python
v_orth = np.array([1, 0])
w_orth = np.array([0, 1])
is_orthogonal = np.dot(v_orth, w_orth) == 0
print(f"Vectors are orthogonal? {is_orthogonal}")
```

## Summary

- Vectors represent data and directions.
- Dot products allow us to calculate angles and projections.
- Orthogonality is key to building efficient, independent features in Machine Learning.
