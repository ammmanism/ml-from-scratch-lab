# 🎲 Hidden Markov Models (HMM) & The Forward-Backward Algorithm

Hidden Markov Models are statistical models where the system being modeled is assumed to be a Markov process with unobservable (i.e., *hidden*) states. They are widely used in temporal pattern recognition, such as speech, handwriting, and bioinformatics.

### Mathematical Foundation
An HMM is defined by:
- $S = \{s_1, s_2, ..., s_N\}$: Set of hidden states.
- $V = \{v_1, v_2, ..., v_M\}$: Set of possible observations.
- $A$: State transition probability matrix.
- $B$: Observation emission probability matrix.
- $\pi$: Initial state distribution.

We aim to implement:
1. **Forward Algorithm**: To compute the probability of an observation sequence.
2. **Viterbi Algorithm**: To find the most likely sequence of hidden states.
3. **Baum-Welch (EM)**: To learn the parameters ($A, B, \pi$) from data.

```python
import numpy as np
import matplotlib.pyplot as plt
from ml_from_scratch.advanced import HiddenMarkovModel

# Set random seed for reproducibility
np.random.seed(42)

plt.style.use('ggplot')
```

## 1. Defining a True HMM (The Weather Toy Example)
Let's imagine a system where the hidden states are the *Weather* (0 = Sunny, 1 = Rainy), and the observations are what a person *Wears* (0 = T-shirt, 1 = Coat, 2 = Umbrella).

```python
true_pi = np.array([0.6, 0.4])  # Start mostly sunny
true_A = np.array([[0.8, 0.2],   # Sunny today -> 80% Sunny tomorrow
                   [0.3, 0.7]])  # Rainy today -> 70% Rainy tomorrow

true_B = np.array([[0.6, 0.3, 0.1],   # Sunny -> T-shirt (60%), Coat (30%), Umbrella (10%)
                   [0.05, 0.25, 0.7]]) # Rainy -> T-shirt (5%), Coat (25%), Umbrella (70%)
```

## 2. Testing the Viterbi Algorithm
If we observe a sequence: [T-shirt, Coat, Umbrella, Umbrella], what is the most likely weather sequence?

```python
hmm = HiddenMarkovModel(n_states=2, n_obs=3)
hmm.pi = true_pi
hmm.A = true_A
hmm.B = true_B

observations = np.array([0, 1, 2, 2]) # T-shirt, Coat, Umbrella, Umbrella
most_likely_states = hmm.viterbi(observations)

weather_map = {0: 'Sunny', 1: 'Rainy'}
print(f"Observations: {observations}")
print(f"Inferred Weather: {[weather_map[s] for s in most_likely_states]}")
```
