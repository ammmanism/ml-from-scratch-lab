# 📈 Experiment 02: Learning Curves and Data Asymptotics

In this experiment, we explore the question: *Do we need a more complex model, or just more data?*

By plotting the Training Error and Validation Error against the number of training samples ($m$), we can diagnose if a model suffers from High Bias (Underfitting) or High Variance (Overfitting).

### Theory
- **High Bias**: Both errors plateau at an unacceptably high level. Adding data won't help; we need a more complex architecture.
- **High Variance**: Training error is low, but validation error is high. There is a large gap. Adding more data *will* help bridge the gap.

```python
import numpy as np
import matplotlib.pyplot as plt
from ml_from_scratch.trees import DecisionTreeRegressor
from ml_from_scratch.linear import LinearRegression

np.random.seed(42)
plt.style.use('ggplot')
```

## 1. Defining the Synthetic Task
We create a highly non-linear target. A simple Linear Regression model will suffer from High Bias.

```python
def generate_data(m):
    X = np.random.uniform(-5, 5, (m, 1))
    y = np.sin(X).ravel() + np.random.normal(0, 0.2, m)
    return X, y

X_train, y_train = generate_data(500)
X_val, y_val = generate_data(200)
```

## 2. Generating the Learning Curve Function

```python
def plot_learning_curves(model, name):
    train_errors, val_errors = [], []
    sample_sizes = np.linspace(10, len(X_train), 20, dtype=int)
    
    for m in sample_sizes:
        model.fit(X_train[:m], y_train[:m])
        
        train_preds = model.predict(X_train[:m])
        val_preds = model.predict(X_val)
        
        train_errors.append(np.mean((train_preds - y_train[:m])**2))
        val_errors.append(np.mean((val_preds - y_val)**2))
        
    plt.plot(sample_sizes, np.sqrt(train_errors), "r-+", linewidth=2, label="Train MSE")
    plt.plot(sample_sizes, np.sqrt(val_errors), "b-", linewidth=3, label="Val MSE")
    plt.legend()
    plt.title(f"Learning Curve: {name}")
    plt.xlabel("Training Set Size")
    plt.ylabel("RMSE")
    plt.ylim([0, 1.5])
    plt.show()
```

## 3. High Bias Regime (Linear Regression)
Notice how both lines plateau at a high error rate. The model cannot capture the sine wave, regardless of how much data we provide.

```python
plot_learning_curves(LinearRegression(method='normal'), "Linear Regression")
```

## 4. High Variance Regime (Unconstrained Decision Tree)
The training error remains near zero, but the validation error is higher. There is a visible gap. Supplying more data (or regularizing via `max_depth`) is needed.

```python
plot_learning_curves(DecisionTreeRegressor(max_depth=None), "Unconstrained Decision Tree")
```
