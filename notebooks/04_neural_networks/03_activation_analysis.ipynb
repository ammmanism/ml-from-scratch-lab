# ⚡ Activation Function Analysis

Exploring the dynamics of ReLU, Sigmoid, and Tanh, and the phenomenon of 'Dead ReLUs'.

```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)

def relu(z):
    return np.maximum(0, z)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def tanh(z):
    return np.tanh(z)

plt.plot(x, relu(x), label='ReLU')
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, tanh(x), label='Tanh')
plt.legend(); plt.grid(True); plt.title('Activation Functions')
plt.show()
```

## 1. The Vanishing Gradient Problem

When using Sigmoid or Tanh in deep networks, the gradients become extremely small for inputs far from zero, causing the lower layers to stop learning. ReLU solves this by having a constant gradient of 1 for all positive inputs.
