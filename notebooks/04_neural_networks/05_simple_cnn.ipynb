# 🖼️ CNN Convolutions from Scratch

Implementing the 2D Convolution operation using NumPy's `sliding_window_view`.

```
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view
import matplotlib.pyplot as plt

def convolve2d(image, kernel):
    # Assumes kernel is smaller than image
    k_h, k_w = kernel.shape
    # Use NumPy to generate a view into the image without copying data
    windows = sliding_window_view(image, (k_h, k_w))
    # Element-wise multiply + sum over the kernel axes
    return np.einsum('ijkh,kh->ij', windows, kernel)

# Dummy image: 10x10
img = np.zeros((10,10))
img[2:8, 2:8] = 1

# Sobel kernel (Vertical Edges)
kernel = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])

out = convolve2d(img, kernel)

plt.subplot(1,2,1); plt.imshow(img, cmap='gray'); plt.title('Original')
plt.subplot(1,2,2); plt.imshow(out, cmap='gray'); plt.title('Filtered (Edges)')
plt.show()
```
