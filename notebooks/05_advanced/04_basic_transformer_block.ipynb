# 🤖 Transformer Self-Attention from Scratch

$Attention(Q, K, V) = softmax(\frac{QK^T}{\sqrt{d_k}})V$

```
import numpy as np

def scaled_dot_product_attention(Q, K, V):
    dk = Q.shape[-1]
    scores = np.matmul(Q, K.swapaxes(-2, -1)) / np.sqrt(dk)
    # Softmax
    weights = np.exp(scores - np.max(scores, axis=-1, keepdims=True))
    weights = weights / np.sum(weights, axis=-1, keepdims=True)
    return np.matmul(weights, V)

# Batch=1, Seq=4, Dim=8
X = np.random.randn(1, 4, 8)
Q = K = V = X # Self-attention

out = scaled_dot_product_attention(Q, K, V)
print(f'Output Shape: {out.shape}')
```

## 1. Multi-Head Attention Intuition

Multi-head attention splits the feature dimension into $h$ heads, allowing the model to attend to different parts of the input sequence simultaneously.
