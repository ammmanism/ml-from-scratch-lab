# Vectors and Matrices in NumPy

This notebook demonstrates fundamental linear algebra operations using NumPy, which are the building blocks of `ml_from_scratch`.

```python
import numpy as np
```

## Vectors
Vectors are 1-D arrays.

```python
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
print(f"v1: {v1}")
print(f"v2: {v2}")
```

```python
# Dot Product
dot_prod = np.dot(v1, v2)
print(f"Dot product: {dot_prod}")
```

## Matrices
Matrices are 2-D arrays.

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
print(f"A:\n{A}")
print(f"B:\n{B}")
```

```python
# Matrix Multiplication
matmul = np.matmul(A, B)
print(f"Matrix Multiplication (A @ B):\n{matmul}")
```
